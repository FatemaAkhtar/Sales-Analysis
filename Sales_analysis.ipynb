<h2>About Data</h2>
Analyze and answer business questions about 12 months worth of sales data. The data contains hundreds of thousands of electronics store purchases broken down by month, product type, cost, purchase address, etc. 

```python
#concatenated the 12 months of data

path = "./Sales_Data"
files = [file for file in os.listdir(path) if not file.startswith('.')] # Ignore hidden files

all_months_data = pd.DataFrame()

for file in files:
    current_data = pd.read_csv(path+"/"+file)
    all_months_data = pd.concat([all_months_data, current_data])
    
all_months_data.to_csv("data.csv", index=False)
```

```python
#dropping NANs
df.dropna(inplace = True)
df.reset_index(drop=True, inplace=True)
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df.dtypes
```

```python
#Changing data types
df = df[df['Quantity Ordered'].str.isdigit()]
df['Quantity Ordered'] = df['Quantity Ordered'].astype('int32')
df['Order ID'] = df['Order ID'].astype('int32')
df['Price Each'] = df['Price Each'].astype('float32')
```

<h2><b>Adding New Attribute</b></h2>

```python
df['Sales'] = df['Price Each'] * df['Quantity Ordered']
```

```python
df['City'] = df['Purchase Address'].str.split(',').apply(lambda x : x[1])
```

```python
a = df['Purchase Address'].str.split(',').apply(lambda x : x[2])
```

```python
df['State'] = a.str.split(' ').apply(lambda x : x[1])
```

```python
df['State'].value_counts()
```

```python
a  = df['Order Date'].str.split(' ').apply(lambda x : x[0])

#Extracting the Month
df['Month'] = a.str.split('/').apply(lambda x : x[0])
```

```python
df['Month'].value_counts()
```

```python
#Extracting The Time

df['Time'] = df['Order Date'].str.split(' ').apply(lambda x : x[1])
```

```python
# Convert string to datetime
df['Time'] = pd.to_datetime(df['Time'])

# Extract hour
#Extract minute
df['Hour'] = df['Time'].dt.hour
df['Minute'] = df['Time'].dt.minute
```

<h3>Q .What was the best month for sales, and how much revenue was generated during that period?</h3>

```python
result = df.groupby('Month')['Sales'].sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 4))
result.plot(kind='bar', color='darkblue',edgecolor = 'black')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation='horizontal')
plt.tight_layout()
plt.show()
```

```python
# Assuming 'result' is the Series containing total sales by month
best_month = result.idxmax()  # Get the index (month) with the maximum sales
revenue = result.max()  # Get the maximum sales revenue

print(f"The best month for sales was {best_month} with a revenue of ${revenue:.2f}")
```

<h3>Q .Which city recorded the highest sales volume?</h3>

```python
result = df.groupby('City')['Sales'].sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 4))
result.plot(kind='pie', autopct='%1.1f%%', startangle=90,shadow=True, colors=plt.cm.tab20.colors)
plt.title('Sales Distribution by City')
plt.axis('equal')  
plt.legend(labels=result.index, loc='center left')
plt.ylabel('')
plt.show()
```

<h3>Q .At what time of day should advertisements be displayed to maximize the likelihood of customers making a purchase?</h3>

<h3>Q.Which products are frequently bought together?</h3>

```python
a = df[df['Order ID'].duplicated(keep=False)]
a['Grouped'] = a.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df2 = a[['Order ID', 'Grouped']].drop_duplicates()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
for i, (key, value) in enumerate(count.most_common(10), 1):
    print(f"{i}. Combination: {' - '.join(key)}, Frequency: {value}")
```

<h3>Q.Which product had the highest sales volume, and what factors contributed to its success?</h3>

```python
product_group = df.groupby('Product')['Quantity Ordered'].sum()
```

```python

product_group_sorted = product_group.sort_values(ascending=False)
plt.figure(figsize=(6, 4))
product_group_sorted.plot(kind='bar',edgecolor = 'black', color='darkgreen')
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.title('Quantity Ordered of Each Product')
plt.show()
```

```python
product_group = df.groupby('Product')
quantity_ordered = product_group['Quantity Ordered'].sum()

keys = [pair for pair, df in product_group]
```

```python
prices = df.groupby('Product')['Price Each'].mean()

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys, quantity_ordered, color='brown', label='Quantity Ordered', edgecolor='black')
ax2.plot(keys, prices, color='black', label='Price ($)', marker='o', markeredgecolor='black')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='brown')
ax2.set_ylabel('Price ($)', color='darkgreen')
ax1.set_xticklabels(keys, rotation='vertical', size=8)

fig.show()
```

<h2><b>Inferences</b></h2>

Upon analyzing the relationship between the price of a product and its quantity ordered, we observed that lower-priced products tend to exhibit higher quantities ordered, indicating greater demand due to their affordability. In contrast, higher-priced products may have lower quantities ordered but contribute significantly to revenue due to their higher prices. This suggests that while lower-priced products drive volume sales, higher-priced products play a crucial role in revenue generation, highlighting the importance of pricing strategies in balancing demand and profitability."
